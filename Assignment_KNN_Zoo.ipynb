```python
# enocde all categorical values to o and 1
#get all categorical columns
cat_columns = zoo.select_dtypes(['object']).columns

#convert all categorical columns to numeric
zoo[cat_columns] = zoo[cat_columns].apply(lambda x: pd.factorize(x)[0])
```

```python
zoo
```

```python
array = zoo.values
X = array[:, 0:17]
Y = array[:, 17]
```

```python
#standardize independent data
scale=StandardScaler()
scaled_data=scale.fit_transform(X)
scaled_data
```

```python
X=scaled_data
X
```

```python
Y
```

```python
kfold = KFold(n_splits=10)
```

```python
model = KNeighborsClassifier(n_neighbors=17)
results = cross_val_score(model, X, Y, cv=kfold)
import warnings
warnings.filterwarnings("ignore")
```

```python
print(results.mean())
```

# Grid Search for Algorithm Tuning

```python
# Grid Search for Algorithm Tuning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# test with 80 neighbours
import numpy
n_neighbors = numpy.array(range(1,81))
param_grid = dict(n_neighbors=n_neighbors)
param_grid
```

```python
KNN = KNeighborsClassifier()
grid = GridSearchCV(estimator=KNN, param_grid=param_grid)
grid.fit(X, Y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

 - we get best score of 98% with 1 neighbour
    
    

# Visualizing the CV results

```python
import matplotlib.pyplot as plt 
%matplotlib inline
# choose k between 1 to 41
k_range = range(1, 81)
k_scores = []
'''use iteration to caclulator different k in models, 
then return the average accuracy based on the cross validation'''
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=5)
    k_scores.append(scores.mean())
# plot to see clearly
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

 - The graph also shows that best results are achcieve with 1 neigbour, with increased neigbhour the score starts declinining.
